```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
```

```python
covid = pd.read_csv('COVID19_state.csv')
```

```python
covid.head()
```

```python
xpos = np.arange((len(covid['State'])))
```

```python
plt.figure(figsize=(20,5))
a1 = plt.bar(xpos,covid['Tested'], width = 0.8,color = 'blue', label = 'Number of testing')
b1 = plt.bar(xpos,covid['Infected'], width = 0.8,color = 'red', label = 'Number of Active Cases')
plt.xticks(xpos,covid['State'],rotation = 90)
plt.legend()
def insert_data_labels(bars):
	for bar in bars:
		bar_height = bar.get_height()
		plt.annotate('{0:.0f}'.format(bar.get_height()),
			xy=(bar.get_x() + bar.get_width() / 2, bar_height),
			xytext=(0, 3),
			textcoords='offset points',
			ha='center',
			va='bottom'
		)

insert_data_labels(a1)
#insert_data_labels(b1)
        
```

```python
plt.figure(figsize=(20,5))
plt.bar(xpos,covid['Deaths'], width = 0.8,color = 'red',  bottom = None,label = 'Number of deaths')
plt.xticks(xpos,covid['State'],rotation = 90)
plt.legend()
```

```python
plt.figure(figsize = (12,5))
plt.scatter(xpos,covid['ICU Beds'], label = 'Number of beds')
plt.title("State and number of ICU Beds ")
plt.xticks(xpos,covid['State'],rotation = 90)
plt.xlabel("State")
plt.ylabel("No of Beds")
plt.show()
plt.legend()
sns.despine
```

```python
plt.figure(figsize = (12,5))
plt.scatter(xpos,covid['Hospitals'], marker = '*', label ='number of hospitals')
plt.title("State and number of Hospitals ")
plt.xticks(xpos,covid['State'],rotation = 90)
plt.xlabel("State")
plt.ylabel(" No of Hospitals")
plt.legend()
plt.show()
sns.despine
```

```python
plt.figure(figsize= (5,5))
plt.title("Relation between infections and temperature")
plt.plot(covid['Temperature'],covid['Infected'],'-o',label ='Infections');
plt.grid()
plt.xlabel('Temperature (in farenheit)')
plt.ylabel('Number of Infections')
plt.legend()
```

```python
plt.figure(figsize = (40,8))
plt.bar(xpos,covid['Age 0-25'], width = 0.3, label ='Age 0-25', color = 'blue')
new_xpos=[]
for item in xpos:
    new_xpos.append(item+0.3)
                    

plt.bar(new_xpos,covid['Age 26-54'],width = 0.3, color = 'yellow', label = 'Age 26-54')
newest_xpos = []
for item in new_xpos:
    newest_xpos.append(item + 0.3)
                    

plt.bar(newest_xpos,covid['Age 55+'],width = 0.3, color ='orange' , label = 'Age 55+')

plt.xticks(xpos,covid['State'],rotation = 90 )


plt.show()
plt.legend()
```
